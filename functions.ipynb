```python
## Program to add two positive numbers numbers


n1,n2 = input('Enter two numbers: ').split(' ')            # spliting string into list then unpacking
flag1= False
flag2 = False
### validate if numbers are valid positive numbers

### n1
if n1.isdigit()| n1.replace('.','',1).isdigit():  # check for valid positive number
    if n1.isdigit():
        n1 = int(n1)
    else:
        n1 = float(n1)
else:
    print('Invalid n1 type')
    flag1 = True

### n2
if n2.isdigit()| n2.replace('.','',1).isdigit():  # check for valid positive number
    if n2.isdigit():
        n2 = int(n2)
    else:
        n2 = float(n2)
else:
    print('Invalid n2 type')
    flag2 = True        
        
### display sum of numbers:
if (flag1==False) & (flag2 == False):
    print('{} + {} = {}'.format(n1,n2, n1+n2))
else:
    print('Error: Invalid Inputs!!')
```

```python
# WAF to find factorial of a number.

def factorial(n):
    if n  < 0:
        return "negative number is invalid"
    else:
        if (n == 0)|(n == 1):
            return 1
        else:
            f = 1
            while n > 1:
                f *= n
                n -= 1
            return f 
        
print(factorial(-3))
print(factorial(0))
print(factorial(1))
print(factorial(3))
print(factorial(5))
```

```python
# WAF to simulate basic calculator program. (+,-,*,/): def calculator(n1,n2,op)

def calculator(n1,n2,op):
    if op == '+':
        return n1+n2
    elif op == '-':
        return n1-n2
    elif op == '*':
        return n1*n2
    elif op == '/':
        if n2 == 0:
            return "Error: division by zero"
        else:
            return n1/n2
    else:
        return "Error: Unknown!"
    
print(calculator(3,4,'+'))
print(calculator(3,4,'/'))
print(calculator(3,0,'/'))
print(calculator(3,4,'plus'))
```

### Dynamic arguments : These arguments can take any number of values in it  
### one,two,five,n ---> any number of arguments passed can be taken by single dynamic argument  

### Dynamic arguments syntax ----> *(identifier) ----> *a
### They are non-keyworded ---> you cannot pass like (a=5,7,6) ---> only values  need to be passed

### when you pass (5,7,9,2) -----> It  takes it as a Tuple & pass to function one by one

```python
# sum()

print(sum([1,2,3,4,5]))                 # Takes iterable and start value as arguments and adds them
print(sum([1,2,3,4,5],50))              # Default start value is 0
```

```python
# WAF to add any number of arguments passed

def add(* n):                # n is dynamic argument ---> all inputs are taken in n as tuple
    return sum(n)
print(add())
print(add(2))
print(add(3,4))
print(add(4,5,6))
print(add(4,5,6,6))
```

```python
def My_fun(* n):
    return n
print(My_fun())               # See n is nothing but a tuple
print(My_fun(2))
print(My_fun(3,4))
print(My_fun(4,5,6))
print(My_fun(4,5,6,6))
```

## To avoid such conditions you can create functions and give inputs as arguments....instead of typecasting

```python
#Given a list in Python and provided the positions of the elements, write a program to swap the two elements in the list.

#Input : List = [23, 65, 19, 90], pos1 = 1, pos2 = 3
#Output : [19, 65, 23, 90]

def swapPositions(lst, pos1, pos2):
     
    lst[pos1], lst[pos2] = lst[pos2], lst[pos1]    # See swaping in python is so easy...you dont need an extra variable
    return lst
 
lst = [23, 65, 19, 90]
pos1, pos2  = 1, 3
 
print(swapPositions(lst, pos1-1, pos2-1))
```

```python
# Given a list of numbers, the task is to write a Python program to find the second largest number in given list
# Input: list1 = [10, 20, 4]
# Output: 10                             # without using sort()

def Maxi(lst1):
    s=lst1[0]
    for i in lst1:
        if i>s:
            s=i
    return s       

lst1=eval(input('Enter your list :')) 
lst1.remove(Maxi(lst1))
Maxi(lst1)
```

```python
#The provided code stub will read in a dictionary containing key/value pairs of name:[marks] for a list of students.
#Print the average of the marks array for the student name provided, showing 2 places after the decimal.

#Sample Input 

#3
#Krishna 67 68 69
#Arjun 70 98 63
#Malika 52 56 60
#Malika

#Sample output
#56.00

n=int(input())
Record={}                               # Used dictionary
for i in range(n):
    x=input().split()
    name,*b =x                          # Used *x
    Record[name]=b
keyy=input()
val=(Record[keyy])
for i in range(len(val)):
    val[i]=float(val[i]) 
ans=(sum(val)/len(val))
print ('%.2f'%ans)                       # Print format for printing upto 2 decimal points
```
